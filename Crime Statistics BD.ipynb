```python
df = df.dropna()
```

```python
#Total Crime Count per Category
crime_cols = df.columns[1:-2]
df[crime_cols] = df[crime_cols].apply(pd.to_numeric, errors='coerce').fillna(0)
crime_sums = df[crime_cols].sum().sort_values(ascending=True)

sns.barplot(
    x=crime_sums.index,
    y=crime_sums.values,
    hue=crime_sums.index,
    palette="flare",
    dodge=False
)

plt.title("Total Crime Count per Category (June 2025)", fontsize=16, weight='bold')
plt.xlabel("Crime Category", fontsize=12)
plt.ylabel("Number of Cases", fontsize=12)
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```

```python
#Area-wise Total Cases
crime_cols = df.columns[1:-1]
df[crime_cols] = df[crime_cols].apply(pd.to_numeric, errors='coerce').fillna(0)
df['Total Cases'] = df[crime_cols].sum(axis=1)

filtered_df = df.iloc[:-1]

sorted_df = filtered_df.sort_values(by='Total Cases', ascending=True)

plt.figure(figsize=(14, 6))

sns.barplot(
    x='Total Cases',
    y='Names of Unit',
    data=sorted_df,
    hue='Names of Unit',
    dodge=False,
    palette='viridis',
    legend=False,
    edgecolor='black'
)


for index, value in enumerate(sorted_df['Total Cases']):
    plt.text(value + 1, index, int(value), va='center', fontsize=10)

plt.title("Area-wise Total Cases", fontsize=16, weight='bold')
plt.xlabel("Total Cases", fontsize=12)
plt.ylabel("Area", fontsize=12)

plt.tight_layout()
plt.show()
```

```python
#Top 5 Crimes
top5 = crime_sums.head(5)
plt.figure(figsize=(8,5))
top5.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title("Top 5 Most Frequent Crimes")
plt.ylabel("")
plt.tight_layout()
plt.show()
```

```python
#Heatmap: Crime Intensity by Unit
crime_df = df.set_index('Names of Unit')[crime_cols]
plt.figure(figsize=(16,8))
sns.heatmap(crime_df, annot=False, cmap='Reds')
plt.title("Heatmap of Crime Intensity by Unit")
plt.xlabel("Crime Category")
plt.ylabel("Unit")
plt.tight_layout()
plt.show()
```
